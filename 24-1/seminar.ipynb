# 대학생활세미나 테이블 정리

```python
import pdfplumber as pp
import pandas as pd
```

```python
#pdf 가져오기
o_pdf = pp.open("24-1(0129).pdf") 
o_pdf
```

```python
#pdf의 각 페이지마다 모든 테이블들을 데이터프레임으로 저장하기
##대생세 강의 테이블은 1, 2(1)페이지에 있음
p0 = o_pdf.pages[0].extract_tables() #첫페이지 모든 테이블 p0에 저장
p0s0 = pd.DataFrame(p0[0]) #첫페이지 테이블 중 첫번째 테이블 p0s0에 저장

p1 = o_pdf.pages[1].extract_tables() #둘째페이지 모든 테이블 p1에 저장
p1s0 = pd.DataFrame(p1[0]) #둘째페이지 테이블 중 첫번째 테이블 p1s0에 저장

print(type(p0s0)) #과정 점검
```

```python
# 열이 동일하기에 합치기
seminar = pd.concat([p0s0, p1s0]) 
# []안에 프레임or시리즈로 전달하기 
#기본으로 axis=0이 적용되어 행방향으로 데이터프레임 연결
seminar
```

```python
#첫 행을 컬럼으로 설정 & 필요없는 행 제거
seminar = seminar.rename(columns=seminar.iloc[0]) #첫행을 컬럼(변수)로 지정
seminar = seminar.iloc[2:] #첫행 제외 나머지 행을 데이터(값/value)로 지정
```

```python
seminar
```

```python
#교과목명에 확인사항 내용 추가하기 
seminar['교과목명'] = seminar['교과목명']+'('+seminar['확인사항']+')'
```

```python
seminar
```

```python
#필요한 컬럼만 남기기
seminar = seminar.loc[:, ['이수\n구분','과목코드','교과목명','교수명','강의실/시간']]
seminar
```

```python
#엉킨 행인덱스 정리
seminar.reset_index(drop=True, inplace=True)
seminar
```

4. 일주일 두번 수업 나누기

```python
df['roomNtime'] = df['roomNtime'].str.replace(" ", "", ) #공백문자 제거. 세 번째 매개변수 비워두면 모든 행의 공백 제거
```

```python
# 일주일 두 번 수업의 특징 : '/n' 존재 -> 이거 기준으로 자르기. 순서 지키기

df['roomNtime2'] = df.roomNtime.str.split('\n').str[1] # '\n' 기준으로 자른것의 두 번째 값
df['roomNtime'] = df.roomNtime.str.split('\n').str[0]  #'\n' 기준으로 자른것의 첫 번째 값
df
# '/n'없는 수업은 room_time2에 nan
```

# 열 변수가 2열짜리인것끼리 처리

```python
import pdfplumber as pp
import pandas as pd
```

```python
pdf = pp.open('24-1(0129).pdf')
pdf
```

```python
frame_list = [] #table 담을 리스트
```

```python
for page in range(4, 11): #사봉 제외한 나머지 교양 중 열변수가 같은 애들끼리 모으기
    tables = pdf.pages[page].extract_tables()
    table_len = len(tables)
    
    for table_n in range(table_len):
        table = tables
        data_frame = pd.DataFrame(table[table_n])
        frame_list.append(data_frame) #테이블 하나씩 frame에 담기
```

```python
len(frame_list)
```

```python
#첫테이블은 사봉이라서 제거
frame_list = frame_list[1:len(frame_list)-1]
len(frame_list)
```

```python
frame_list
```

```python
# 열변수 지정
```

```python
type(frame_list[0])
```

```python
# 필요없는 열변수 제거
```

```python
for n in range(len(frame_list)):
    frame = frame_list[n].rename(columns=frame_list[n].iloc[0]) #첫행을 컬럼(변수)로 지정
    frame = frame.iloc[2:] #3번째 행부터 데이터(값)으로 지정
    frame_list[n] = frame[['이수\n구분', '과목코드', '교과목명', '교수명', '강의실/시간']]
```

```python
# 데이터 확인 : 순번으로...
```

```python
# 모든 테이블하나로 합치기
```

```python
df = pd.concat(frame_list)
df
```

```python
# 엉킨 인덱스 정리
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('csv3.csv', index=False)
```

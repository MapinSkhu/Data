# 말과글, 인권과평화, 과에콜, 데활분 테이블 정리

```python
import pdfplumber as pp
import pandas as pd
```

```python
pdf = pp.open("24-1(0129).pdf")
pdf
```

```python
# 정리할 테이블은 2(2), 3, 4(1)에 있음
p1 = pdf.pages[1].extract_tables()
p1s1 = pd.DataFrame(p1[1]) #말과글1

p2 = pdf.pages[2].extract_tables()
p2s0 = pd.DataFrame(p2[0]) #말과글2(변수 없음)
p2s1 = pd.DataFrame(p2[1]) #인평
p2s2 = pd.DataFrame(p2[2]) #과에콜

p3 = pdf.pages[3].extract_tables()
p3s0 = pd.DataFrame(p3[0]) #데활

type(p1s1)
```

```python
p1s1_new = pd.concat([p1s1, p2s0])
```

```python
frame_list = [p1s1_new, p2s1, p2s2, p3s0]
```

```python
for n in range(len(frame_list)):
    frame = frame_list[n].rename(columns=frame_list[n].iloc[0]) #첫행을 컬럼(변수)로 지정
    frame_list[n] = frame.iloc[2:] #3번째 행부터 데이터(값)으로 지정
```

```python
frame_list[0]
```

```python
df = pd.concat(frame_list) #열(변수)가 같으니 열에 맞게 값이 합쳐졌다
```

```python
df
```

```python
#필요한 열만 남기기
df = df[['이수\n구분', '과목코드', '교과목명', '교수명', '강의실/시간']]
df
```

```python
# 엉킨 인덱스 정리
df.reset_index(drop=True, inplace=True)
df
```

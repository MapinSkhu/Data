## 말과글, 인평, 과에콜, 데활분 테이블 생성

1. 연장 준비

```python
import pdfplumber as pp

import pandas as pd
```

```python
o_pdf = pp.open("23-2.pdf") 
o_pdf
```

2. 정리할 테이블 추출&데이터프레임으로 저장

```python
p2 = o_pdf.pages[2].extract_tables()
w = pd.DataFrame(p2[0])
h = pd.DataFrame(p2[1])

p3 = o_pdf.pages[3].extract_tables()
s = pd.DataFrame(p3[0])
d = pd.DataFrame(p3[1])

print(type(w))
```

3. 한 번에 처리하기 위해 리스트에 테이블 담기

```python
required = [w, h, s, d]
required
```

4. 필요한 열만 남기기 & 첫 번째 테이블 제외한 나머지 헤더 삭제

```python

for n in range(len(required)):
    t = required[n]
    t = t.loc[:, [5,7,8,9,11,12]]
    required[n] = t.drop(t.index[1])
for n in range(len(required)-1):
    t = required[n+1]
    required[n+1] = t.drop(t.index[0])

required
```

5. 테이블 하나로 합치기

```python
required = pd.concat(required)
required
```

```python
type(required)
```

6. 테이블 헤더 지정 & 엉킨 인덱스 초기화

```python
df = required.copy()

df_column = df.iloc[0]
df = df[1:]
df.columns = df_column

df.reset_index(drop=True, inplace=True) #추가 설정 : 기존 인덱스 삭제, 데이터프레임에 바로 적용
df
```

# 채플 테이블 생성

1. 연장 준비

```python
import pdfplumber as pp

import pandas as pd
```

```python
o_pdf = pp.open("23-2.pdf") 
o_pdf
```

2. 정리할 테이블 추출

```python
p4 = o_pdf.pages[4].extract_tables() #4번째 페이지에 있는 모든 테이블 추출
ch = pd.DataFrame(p4[0]) #추출한 테이블 중 1번째 테이블 추출

ch
```

3. 불필요한 열 삭제

```python
ch = ch.loc[:, [5,7,8,9,11,12,13]]
ch
```

4. 불필요한 행 삭제

```python
ch = ch.drop(ch.index[1])
ch
```

5. 헤더 설정 & 인덱스 초기화

```python
df = ch.copy()
```

```python
df_column = df.iloc[0]
df = df[1:]
df.columns = df_column

df.reset_index(drop=True, inplace=True)
df
```

6. 필요한 대로 값 변경

```python
df['교과목명'] = df['교과목명'] +'('+ df['세부내용'] +')'
df = df.drop(['세부내용'], axis=1)
df
```

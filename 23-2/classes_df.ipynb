# 정리한 ori 테이블 모두 합치기

```python
import pandas as pd
```

1. ori 테이블 불러오기

```python
chapel = pd.read_csv('ori-chapel.csv')
major = pd.read_csv('ori-major.csv')
required = pd.read_csv('ori-required.csv')
seminar = pd.read_csv('ori-seminar.csv')
```

```python
type(chapel)
```

2. 테이블 변수 이름 맞추기

```python
chapel.columns = ['kind','code','num','class_name','prof','room_time']
required.columns = ['kind','code','num','class_name','prof','room_time']
seminar.columns = ['kind','code','num','class_name','prof','room_time']
```

```python
#확인
chapel
```

3. 합치기

```python
concat_class = pd.concat([seminar, required, chapel, major])
```

```python
concat_class.reset_index(drop=True, inplace=True)
concat_class
```

4. 필요한대로 변수 정리

4-1. 일주일 두 번 수업 분리

```python
df = concat_class.copy()
```

```python
# 작업 편하게 room_time 값에서 공백 제거

df['room_time'] = df['room_time'].str.replace(" ", "", ) #공백문자 제거. 세 번째 매개변수 비워두면 모든 행의 공백 제거
```

```python
# 일주일 두 번 수업의 특징 : '/n' 존재 -> 이거 기준으로 자르기. 순서 지키기

df['room_time2'] = df.room_time.str.split('\n').str[1] # '\n' 기준으로 자른것의 두 번째 값
df['room_time'] = df.room_time.str.split('\n').str[0]  #'\n' 기준으로 자른것의 첫 번째 값
df
# '/n'없는 수업은 room_time2에 nan
```

4-2. room_time에서 room, date1, date2, start, end 분리

```python
df2 = df.copy()
```

```python
type(df2.room_time2[0])
```

```python
df2['room'] = df2.room_time.str.split('/').str[0]

df2['date1'] = df2.room_time.str.split('/').str[1]
df2['date2'] = df2.room_time2.str.split('/').str[1]

df2['time'] = df2.room_time.str.split('/').str[2]
df2['start'] = df2.time.str.split('~').str[0]
df2['end'] = df2.time.str.split('~').str[1]
```

```python
df2
```

```python
# 필요한 속성만 남기기
class_df = df2[['kind', 'code', 'class_name', 'prof', 'room', 'date1', 'date2', 'start', 'end']]
class_df
```

4-3. kwan 변수 추가

```python
import numpy as np
```

```python
kwan_df = class_df.copy()
```

```python
kwan_df['kwan'] = kwan_df['room'].str[0]

conditionlist = [
    (kwan_df['kwan'] == '1'),
    (kwan_df['kwan'] == '3'), 
    (kwan_df['kwan'] == 'M'),
    (kwan_df['kwan'] == '9'),
    (kwan_df['kwan'] == '7'),
    (kwan_df['kwan'] == '6'),
    (kwan_df['kwan'] == 'B'),
    (kwan_df['kwan'] == '5')
] #첫 문자로 관 판별
choicelist = [
    '승연관',
    '월당관',
    '미가엘관',
    '성미가엘성당&피츠버그홀',
    '새천년관',
    '정보과학관',
    '일만관',
    '나눔관'
]

kwan_df['kwan'] = np.select(conditionlist, choicelist, default='미정') # room이 미정인거 처리
```

```python
kwan_df
```

4-4. room == 9999 삭제

```python
index_9999 =kwan_df[kwan_df['room']=='9999'].index
```

```python
kwan_df.loc[index_9999]
```

```python
kwan_df.drop(index_9999, inplace=True)
kwan_df.reset_index(drop=True, inplace=True)
```

```python
kwan_df
```

5. 요일순 정렬

```python
sort_df = kwan_df.copy()
```

```python
sort_df.sort_values(by=['kwan', 'room', 'date1', 'date2', 'start', 'end', 'class_name'], inplace=True) # 맵인스쿠 함수 작동에 편리하게 정렬
```

```python
sort_df
```

## 대생세 테이블 생성

```python
import pdfplumber as pp

import pandas as pd
```

```python
o_pdf = pp.open("23-2.pdf") 
o_pdf #pdf 한페이지씩 돌면서 각 페이지 테이블 추출해야함
```

```python
data = [] #테이블 저장할 리스트
```

```python
s1 = o_pdf.pages[0].extract_tables()
s1 = pd.DataFrame(s1[0])

s2 = o_pdf.pages[1].extract_tables()
s2 = pd.DataFrame(s2[0])

print(type(s1))
```

```python
seminar = pd.concat([s1, s2]) #기본으로 axis=0이 적용되어 행방향으로 데이터프레임 연결
seminar
```

```python
seminar2 = seminar.loc[:, [5,7,8,9,11,12,13]]
seminar2
```

```python
seminar3 = seminar2.drop(seminar2.index[1])
seminar3
```

```python
## 헤더 지정해주기
df = seminar3.copy()
df_column = df.iloc[0]
df = df[1:]
df.columns = df_column

df
```

```python
# 인덱스 초기화
df.reset_index(drop=True, inplace=True) #추가 설정 : 기존 인덱스 삭제, 데이터프레임에 바로 적용
df
```

```python
df['교과목명'] = df['교과목명']+'('+df['확인사항']+')'

df
```

```python
df = df.drop(['확인사항'], axis=1)
df
```

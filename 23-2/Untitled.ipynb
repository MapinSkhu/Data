# 교양필수 제외 나머지 수업 테이블 생성

1. 연장 준비

```python
import pdfplumber as pp
import pandas as pd
```

```python
o_pdf = pp.open('23-2.pdf')
o_pdf
```

2. 필요한 테이블 추출

```python
o_df1 = [] #(pdf상)p6 ~ p7
o_df2 = [] #(pdf상)p8 ~ 끝까지
```

```python
for pn in range(5, 7): #(pdf상)p6 ~ p7
    table_len = len(o_pdf.pages[pn].extract_tables())
    
    for tn in range(table_len):
        pre = o_pdf.pages[pn].extract_tables() 
        o_df1.append(pd.DataFrame(pre[tn])) #테이블 하나씩 데이터프레임으로 변환후 data에 담기
        
for pn in range(7, len(o_pdf.pages)): #(pdf상)p8 ~ 끝까지
    table_len = len(o_pdf.pages[pn].extract_tables())
    
    for tn in range(table_len):
        pre = o_pdf.pages[pn].extract_tables() 
        o_df2.append(pd.DataFrame(pre[tn])) #테이블 하나씩 데이터프레임으로 변환후 data에 담기
```

```python
type(o_df1[0])
```

```python
print(len(o_df1), len(o_df2))
```

3. 필요한 열/행만 남기기

```python
# o_df1 정리
o_df1[0]
```

```python
o_df1[0] = o_df1[0].loc[:, [1,3,4,5,7,9]] # 첫 번째 테이블 제외한 나머지 인덱스0은 삭제

for tn in range(len(o_df1)-1):
    t = o_df1[tn+1].loc[:, [1,3,4,5,7,9]]
    o_df1[tn+1] = t.drop(t.index[0])
    
o_df1
```

4. 정리된 테이블 연결하기

```python
df1 = o_df1.copy()
```

```python
df1 = pd.concat(df1)
```

5. 정리 중인 테이블 헤더 지정

```python
df1_column = df1.iloc[0]
df1 = df1[1:]
df1.columns = df1_column
```

```python
t1 = df1.copy()
t1
```

# 모든 테이블을 데이터프레임 형태로 저장

```python
tlist = [] #미리 공간할당하고 싶으나 좋은방법 찾지 못함
```

```python
for i in range(len(pdf.pages)-1): #마지막장 제외
    page_table = pdf.pages[i].extract_tables()
    for j in range(len(page_table)):
        tlist.append(pd.DataFrame(page_table[j]))
```

# 필요한  열만  남기기

```python
# 첫행을 컬럼으로 지정
for index, df in enumerate(tlist):
    data = df.rename(columns=df.iloc[0]) #첫행을 컬럼(변수)로 지정
    data = data.iloc[1:] #첫행 제외 나머지 행을 데이터(값/value)로 지정
    tlist[index] = data
```

# 강의실, 시간, 요일 컬럼 나누기

```python
# rnt 공백문자 제거
table['rnt'] = table['rnt'].str.replace(" ", "", ) #공백문자 제거. 세 번째 매개변수 비워두면 모든 행의 공백 제거
table[375:]
```

```python
# 주2회 수업 특징 : '/n' 존재 -> 이거 기준으로 자르기. 순서 중요

table['rnt2'] = table.rnt.str.split('\n').str[1] # '\n' 기준으로 자른것의 두 번째 값
table['rnt'] = table.rnt.str.split('\n').str[0] # '\n' 기준으로 자른것의 첫 번째 값
table
# '/n'없는 수업은 room_time2에 nan
```

```python
# rnt/rnt2의 강의실,시간이 서로 동일한지 확인
deleted_nan = table.dropna(axis=0, how='any', subset=['rnt2'], inplace=False).copy()# rnt2가 nan인 데이터 제외시키기
# COPY 안하면 경고뜸
```

```python
deleted_nan
```

```python
deleted_nan['test_room'] = (deleted_nan.rnt.str.split('/').str[0] != deleted_nan.rnt2.str.split('/').str[0])
deleted_nan['test_time'] = (deleted_nan.rnt.str.split('/').str[2] != deleted_nan.rnt2.str.split('/').str[2])
```

```python
deleted_nan
```

```python
# 주 2회 수업 중 강의실과 시간이 다른 행 추출
deleted_nan[deleted_nan['test_room']==True]
```

```python
deleted_nan[deleted_nan['test_time']==True]
```

```python
# 강의실, 강의시간, 강의요일 분리
table['room'] = table.rnt.str.split('/').str[0] 
table['date1'] = table.rnt.str.split('/').str[1] 
table['date2'] = table.rnt2.str.split('/').str[1] 
table['start'] = table.rnt.str.split('/').str[2].str.split('~').str[0]
table['end'] = table.rnt.str.split('/').str[2].str.split('~').str[1]
table
```

# 관 컬럼추가 

```python
import numpy as np
```

```python
ktable = table.copy()
```

```python
ktable['kwan'] = ktable['room'].str[0]

conditionlist = [
    (ktable['kwan'] == '1'),
    (ktable['kwan'] == '3'), 
    (ktable['kwan'] == 'M'),
    (ktable['kwan'] == '9'),
    (ktable['kwan'] == '7'),
    (ktable['kwan'] == '6'),
    (ktable['kwan'] == 'B'),
    (ktable['kwan'] == '5')
] #첫 문자로 관 판별
choicelist = [
    '승연관',
    '월당관',
    '미가엘관',
    '성미가엘성당&피츠버그홀',
    '새천년관',
    '정보과학관',
    '일만관',
    '나눔관'
]

ktable['kwan'] = np.select(conditionlist, choicelist, default='미정') 
```

```python
ktable
```

```python
table = ktable.copy()
table
```

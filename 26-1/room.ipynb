# 관 컬럼 추가

```python
room['kwan'] = room['room'].str[0]

conditionlist = [
    (room['kwan'] == '1'),
    (room['kwan'] == '3'),
    (room['kwan'] == 'M'),
    (room['kwan'] == '9'),
    (room['kwan'] == '7'),
    (room['kwan'] == '6'),
    (room['kwan'] == 'B'),
    (room['kwan'] == '5')
] #첫 문자로 관 판별
choicelist = [
    '승연관',
    '월당관',
    '미가엘관',
    '성미가엘성당&피츠버그홀',
    '새천년관',
    '정보과학관',
    '일만관',
    '나눔관'
]

room['kwan'] = np.select(conditionlist, choicelist, default='미정')
```

```python
room
```

```python
room['room_image'] = np.nan
room['room_type'] = np.nan
```

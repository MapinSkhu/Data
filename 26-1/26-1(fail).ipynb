# 데이터 준비

```python
import tabula
import pandas as pd

file_path = "/content/drive/MyDrive/mapinskhu(26~)/Data/26-1.pdf"

# {페이지번호: [가져올 테이블 인덱스 리스트]}
# 예: 1페이지에서는 0번, 1번 테이블 / 2페이지에서는 1번 테이블만 / 3페이지에서는 0번, 2번 테이블
target_config = {
    21 : [0,1],
    22 : [0,1],
    23 : [0],
    24 : [0,1,2],
    25 : [0,1],
    26 : [0,1,2],
    27 : [0,1],
    28 : [0,1,2],
    29 : [0,1],
    30 : [0,1,2],
    31 : [0],
    32 : [0,1],
    33 : [0,1,2],
    34 : [0,1,2],
    39 : [0],
    43 : [0],
    47 : [0],
    51 : [0],
    57 : [0],
    58 : [0],
    63 : [0],
    67 : [0],
    72 : [0],
    77 : [0],
    82 : [0],
    83 : [0],
    89 : [0],
    93 : [0],
    96 : [1],
    99 : [1],
    103 : [0],
    109 : [0],
    113 : [0],
    117 : [0],
    125 : [1],
    126 : [0],
    127 : [0]
}

all_selected_tables = []

for page, indices in target_config.items():
    # 해당 페이지의 모든 테이블을 일단 다 읽어옵니다.
    tables = tabula.read_pdf(file_path, pages=page, multiple_tables=True)

    if not tables:
        print(f"{page}페이지에 추출된 테이블이 없습니다.")
        continue

    for idx in indices:
        if len(tables) > idx:
            # 선택한 테이블을 리스트에 추가
            all_selected_tables.append(tables[idx])
            print(f"{page}페이지의 {idx}번 테이블 추출 성공")
        else:
            print(f"{page}페이지에 {idx}번 인덱스의 테이블이 존재하지 않습니다.")
```

```python
all_selected_tables[0]
```

```python
processed_tables = []

for df in all_selected_tables:

    # 공통 열 추출
    common_columns = ['이수\r구분', '과목코드', '교과목명', '교수명', '강의실/시간', '확인사항']
    available_cols = [col for col in common_columns if col in df.columns]

    df_final = df[available_cols].copy()
    processed_tables.append(df_final)

# 최종 병합
if processed_tables:
    final_df = pd.concat(processed_tables, ignore_index=True)
    print("병합 완료!")
```

```python
# 특정 행만 컬럼끼리 이어 붙이기
# 데이터가 문자열인지 확인하기 위해 astype(str)을 사용합니다.

target_col = '교과목명'  # 값이 '헬로'인지 확인할 컬럼
data_col = '확인사항'    # 가져와서 붙일 내용이 있는 컬럼

# 조건
condition1 = (final_df[target_col] == '인권과평화')
condition2 = (final_df[target_col] == '대학생활세미나')

# 이어 붙이기
final_df.loc[condition1, target_col] = (
    final_df.loc[condition1, target_col].astype(str) + "(" +
    final_df.loc[condition1, data_col].astype(str) + ")"
)
final_df.loc[condition2, target_col] = (
    final_df.loc[condition2, target_col].astype(str) + "(" +
    final_df.loc[condition2, data_col].astype(str) + ")"
)
```

```python
final_df = final_df.drop(columns=['확인사항'])
```

# 데이터 가공

## 주2회 수업 분리

```python
# 값에서 공백제거
df1 = final_df.copy()

# \s+ 는 모든 공백 문자(스페이스, 탭, 줄바꿈 등)를 의미합니다.
# 이를 빈 문자열('')로 치환하면 문자열 내 모든 공백이 사라집니다.
# df1 = df1.replace({r'\s+': ''}, regex=True)

cols_to_fix = ['kind','code','class_name','prof']

for col in cols_to_fix:
    if col in df1.columns:
        # 모든 공백 문자를 찾아 삭제합니다.
        df1[col] = df1[col].astype(str).str.replace(r'\s+', '', regex=True)
```

```python
df2 = df1.copy()
df2['roomntime1'] = df2.roomntime.str.split('\r').str[0]
df2['roomntime2'] = df2.roomntime.str.split('\r').str[1]
df2.drop(columns=['roomntime'], inplace=True)
```

## 강의실, 요일, 시간 분리

```python

# 주2회 수업중 각 수업의 시간과 강의실이 같은지 확인
# rnt2가 nan인 데이터 제외 추출
df3 = df2.dropna(axis=0, how='any', subset=['roomntime2'], inplace=False).copy()
```

```python
# room 또는 time이 일치하면 false 저장
df3['testRoom'] = \
(df3.roomntime1.str.split('/').str[0]) \
!= (df3.roomntime2.str.split('/').str[0])

df3['testTime'] = \
(df3.roomntime1.str.split('/').str[2]) \
!= (df3.roomntime2.str.split('/').str[2])
```

```python
df3.head()
```

```python
column_counts = df3[['testRoom', 'testTime']].sum()

print("--- 컬럼별 True 개수 ---")
print(column_counts)
```
